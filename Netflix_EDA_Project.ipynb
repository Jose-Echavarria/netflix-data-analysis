```python
import pandas as pd
```

```python
data = pd.read_csv("netflix1.csv")
```

```python
data.head(5)
```

```python
data.info()
```

```python
data.columns
```

```python
data.drop(columns = "rating", inplace = True)
```

```python
data
```

```python
new_column_names = []

for i in data.columns:
    new_column_names.append(i.capitalize())
```

```python
data.columns = new_column_names
```

```python
data
```

```python
data.duplicated().sum()
```

```python
# If we had duplicates

data.drop_duplicates(inplace=True)
```

```python
data.isna().sum()
```

```python
#If we had na values

data.dropna(inplace=True)
```

```python
data.head()
```

```python
data["Show_id"].unique()
```

```python
data["Show_id"] = data["Show_id"].apply(lambda x: x.split("s")[1])
```

```python
data
```

```python
type(data["Show_id"][0])
```

```python
data["Show_id"] = data["Show_id"].astype(int)
```

```python
data["Date_added"] = data["Date_added"].apply(lambda x: x.replace("/","-"))
```

```python
data.head(6)
```

```python
data.to_csv("Cleaned_Data_CSV.csv", index = False)
```

```python
data.to_excel("Cleaned_Data_Excel.xlsx", index=False)
```
