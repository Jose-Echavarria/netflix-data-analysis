```python
data["Show_id"] = data["Show_id"].apply(lambda x: x.split("s")[1])
```

```python
data["Show_id"] = data["Show_id"].astype(int)
```

```python
data["Date_added"] = data["Date_added"].apply(lambda x: x.replace("/","-"))
```

```python
data.head(6)
```

```python
data.to_csv("Cleaned_Data_CSV.csv", index = False)
```

```python
data.to_excel("Cleaned_Data_Excel.xlsx", index=False)
```

```python
data["Date_added"] = pd.to_datetime(data["Date_added"], errors='coerce')
data["Year"] = data["Date_added"].dt.year
```

```python
data["Country"] = data["Country"].str.split(", ")
data = data.explode("Country")
```

```python
# Count titles per year per country
content_growth = data.groupby(["Year", "Country"]).size().reset_index(name="Count")

# Optional: focus on top countries by total titles
top_countries = content_growth.groupby("Country")["Count"].sum().nlargest(6).index
filtered = content_growth[content_growth["Country"].isin(top_countries)]

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered, x="Year", y="Count", hue="Country", marker="o")
plt.title("Growth of Netflix Content by Country Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Explode genre
data["Genre"] = data["Listed_in"].str.split(", ")
data = data.explode("Genre")
```

```python
# Count unique genres per country
genre_diversity = data.groupby("Country")["Genre"].nunique().reset_index()
genre_diversity = genre_diversity.sort_values(by="Genre", ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=genre_diversity, x="Genre", y="Country", palette="viridis")
plt.title("Top 10 Countries by Genre Diversity")
plt.xlabel("Unique Genres")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
